```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file by specifying the file path directly
# Replace 'your_file_path.csv' with the actual file path in your Colab environment
file_path = '/content/deu_with_pos_stanza.csv'  # Change this to the actual path of your CSV file
df = pd.read_csv(file_path)

# Ensure the columns are properly named
# Rename columns if necessary (change these names to match your actual file)
df.columns = ['Word', 'Translation', 'Language', 'Part of Speech']

# Drop any rows with missing values in important columns
df.dropna(subset=['Word', 'Part of Speech'], inplace=True)

# 1. Count and percentage of words per part of speech
pos_counts = df['Part of Speech'].value_counts()
pos_percentages = (pos_counts / pos_counts.sum()) * 100

print("Words per Part of Speech:")
print(pos_counts)
print("\nPercentage of Words per Part of Speech:")
print(pos_percentages)

# Plotting Part of Speech Distribution
plt.figure(figsize=(10, 6))
pos_counts.plot(kind='bar')
plt.title('Part of Speech Distribution')
plt.xlabel('Part of Speech')
plt.ylabel('Word Count')
plt.xticks(rotation=45)
plt.show()

# 2. Frequency distribution of word lengths
df['Word Length'] = df['Word'].apply(len)
word_length_counts = df['Word Length'].value_counts().sort_index()

print("\nFrequency Distribution of Word Lengths:")
print(word_length_counts)

# Plotting Word Length Distribution
plt.figure(figsize=(10, 6))
word_length_counts.plot(kind='bar')
plt.title('Frequency Distribution of Word Lengths')
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.show()
```
