```python
#deu pos tagging success

# ✅ Step 1: Install Stanza (Multilingual POS Tagger)
!pip install stanza

# ✅ Step 2: Import Required Libraries
import pandas as pd
import stanza

# ✅ Step 3: Download the German Language Model
stanza.download('de')  # Download the German language model

# ✅ Step 4: Initialize the German NLP Pipeline
nlp = stanza.Pipeline('de')  # Load German language for POS tagging

# ✅ Step 5: Load the CSV (Assumes 2 Columns)
flashcards = pd.read_csv('/content/deu.csv', header=None)
flashcards.columns = ["Word", "Translation"]
flashcards["Language"] = "German"

# ✅ Step 6: Define a Function for German POS Tagging
def get_german_pos(word):
    try:
        # Process the word using the German model
        doc = nlp(word)
        # Extract the Universal POS tag from the processed result
        return doc.sentences[0].words[0].upos  # Returns POS tags like NOUN, VERB, ADJ
    except Exception as e:
        print(f"Error tagging '{word}': {e}")
        return "Unknown"

# ✅ Step 7: Apply POS Tagging Function
flashcards["Part of Speech"] = flashcards["Word"].apply(get_german_pos)

# ✅ Step 8: Display the Results
print(flashcards.head())

# ✅ Step 9: Save the Updated CSV with German POS Tags
flashcards.to_csv('/content/deu_with_pos_stanza.csv', index=False)
print("✅ POS tagging with Stanza completed successfully!")

# ✅ Step 10: Download the Updated File to Your Device
from google.colab import files
files.download('/content/deu_with_pos_stanza.csv')
```
